F**orcasting with FB PROPHET ALGORITHM**

```python
#converting the date from DD/MM/YYYY to YYYY-MM-DD

date_info=pd.to_datetime(air_quality_data['Date'], format='%d/%m/%Y')
print(date_info)
```

```python
time_info=air_quality_data['Time']
```

```python
print(time_info)
```

```python
time_info=time_info.apply(lambda x:x.replace('.',':'))
```

```python
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
#combining 2 series to a pandas dataframe
date_time=pd.concat([date_info,time_info],axis=1)
```

```python
date_time.head()
```

```python
date_time.shape
```

```python
from datetime import date
#combining date and time
date_time['ds']=date_time['Date'].astype(str)+' '+date_time['Time'].astype(str)
```

```python
date_time.head()
```

```python
date_time.info()
```

WE have to convert ds from object datatype to date time **formate**

```python
data=pd.DataFrame()
```

```python
data['ds']=pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y']=air_quality_data['RH']
```

```python
data.head()
```

```python
from prophet import Prophet
```

```python
data['y'] = pd.to_numeric(data['y'], errors='coerce')
```

```python
model=Prophet()
model.fit(data)
```

```python
future=model.make_future_dataframe(periods=365,freq='H')
future.tail()
```

```python
forecast=model.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
fig1=model.plot(forecast)
```

```python
fig2=model.plot_components(forecast)
```

```python
from prophet.plot import plot_plotly, plot_components_plotly

# Interactive forecast plot (better than static)
fig1 = plot_plotly(model, forecast)
fig1.show()
```

```python
# Filter only future dates (after last observed date)
future_forecast = forecast[forecast['ds'] > data['ds'].max()]

fig_future = model.plot(future_forecast)
```
